```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns=np.append(cancer['feature_names'], ['target']))
```

```python
target_feat = df.groupby('target').size().reset_index(name='Count')
plt.figure(figsize=(10, 7))
sns.set_style('whitegrid')
sns.countplot(x='target', data=df)
target_feat
```

```python
sns.pairplot(df, hue='target', vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness'])
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(df.corr(), annot=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('target', axis=1)
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X_train, y_train)
```

```python
predictions = logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```

```python
accuracy = round(logreg.score(X_train, y_train) * 100, 2)
print(f'accuracy: {accuracy}%')
```
